## Analysing Genres

```python
df1 = df.assign(Genres=df['Genres'].str.split(', ')).explode('Genres')
characters_to_exclude = ['[', ']', "'"]
for char in characters_to_exclude:
    df1['Genres'] = df1['Genres'].str.replace(char, '')
df1
```

```python
df1.groupby('Genres').count()
```

```python
genres_counts = df1['Genres'].value_counts()
plt.figure(figsize=(10, 10))
# Create a horizontal bar plot for the top 10 occurrences
plt.barh(genres_counts.index, genres_counts.values)
plt.xlabel('Count')
plt.ylabel('Genres')
plt.title('Occurrences of Genres')
plt.show()





```

```python
type_mean_popularity = df1.groupby('Genres')['Popularity'].mean()

# Increase the size of the figure
plt.figure(figsize=(10, 10))

# Create a horizontal bar plot
plt.barh(type_mean_popularity.index, type_mean_popularity.values)
plt.xlabel('Mean Popularity')
plt.ylabel('Genres')
plt.title('Mean Popularity for Different Genres')
plt.show()
```

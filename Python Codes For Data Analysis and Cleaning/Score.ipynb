```python
dft_=pd.read_csv('anime.csv')
dft=dft_[['MAL_ID','Score','Score-10','Score-9','Score-8','Score-7','Score-6','Score-5','Score-4','Score-3','Score-2','Score-1']].copy()
dft.replace('Unknown', pd.NA, inplace=True)
dft
```

```python
dft['Score-10'] = pd.to_numeric(dft['Score-10'], errors='coerce').fillna(0).astype('int64')
dft['Score-9'] = pd.to_numeric(dft['Score-9'], errors='coerce').fillna(0).astype('int64')
dft['Score-8'] = pd.to_numeric(dft['Score-8'], errors='coerce').fillna(0).astype('int64')
dft['Score-7'] = pd.to_numeric(dft['Score-7'], errors='coerce').fillna(0).astype('int64')
dft['Score-6'] = pd.to_numeric(dft['Score-6'], errors='coerce').fillna(0).astype('int64')
dft['Score-5'] = pd.to_numeric(dft['Score-5'], errors='coerce').fillna(0).astype('int64')
dft['Score-4'] = pd.to_numeric(dft['Score-4'], errors='coerce').fillna(0).astype('int64')
dft['Score-3'] = pd.to_numeric(dft['Score-3'], errors='coerce').fillna(0).astype('int64')
dft['Score-2'] = pd.to_numeric(dft['Score-2'], errors='coerce').fillna(0).astype('int64')
dft['Score-1'] = pd.to_numeric(dft['Score-1'], errors='coerce').fillna(0).astype('int64')
```

```python
dft.info()
```

```python
dft['total_score']=((dft['Score-10']*10)+(dft['Score-9']*9)+(dft['Score-8']*8)+(dft['Score-7']*7)+(dft['Score-6']*6)+(dft['Score-5']*5)+(dft['Score-4']*4)+(dft['Score-3']*3)+(dft['Score-2']*2)+(dft['Score-1']*1))
dft['total_votes']=((dft['Score-10'])+(dft['Score-9'])+(dft['Score-8'])+(dft['Score-7'])+(dft['Score-6'])+(dft['Score-5'])+(dft['Score-4'])+(dft['Score-3'])+(dft['Score-2'])+(dft['Score-1']))
dft['score_new']=(dft['total_score']/dft['total_votes'])

```

```python
dft['score_new'].fillna(0,inplace=True)
dft['Score'].fillna(0,inplace=True)
```

```python
df.set_index('MAL_ID', inplace=True)
df.update(dft['score_new'])
df.reset_index(inplace=True)
```

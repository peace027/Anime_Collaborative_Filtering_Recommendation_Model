## Implementation of the scrapper for Myanimelist website.

```python
df=pd.read_csv('step_3_ImageURL.csv')
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

# Assuming 'df_' is your DataFrame

# Create a function to fetch image URLs for a given MAL_ID
def fetch_image_urls(mal_id):
    time.sleep(1)
    
    url = "https://myanimelist.net/anime/" + str(mal_id) + "/_/pics"

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
    webpage = requests.get(url, headers=headers).text

    soup = BeautifulSoup(webpage, 'lxml')

    img_1 = soup.find_all('img', attrs={'class': 'lazyload'})
    img = []
    for i in img_1:
        img.append(i.get('data-src'))
    
    return img

# Update 'Image URL List' column for the first 100 rows
for i in range(30000,50000):    
    df1 = df__[(df__['MAL_ID'] < (i + 1) * 1) & (df__['MAL_ID'] >= i * 1)].copy()
    df1.reset_index(drop=True, inplace=True)
    df1['Image URL List'] = None
    
    # Update 'Image URL List' column for the first 100 rows
    for j in df1.index:
        if pd.isna(df1.loc[j, 'Image URL List']):
            img_urls = fetch_image_urls(df1.loc[j, 'MAL_ID'])
            if img_urls:  # Check if img_urls is not empty before assigning
                df1.at[j, 'Image URL List'] = img_urls
                print(df1.loc[j, 'MAL_ID'])
            
    df__.set_index('MAL_ID', inplace=True)
    df1.set_index('MAL_ID', inplace=True)

    df__.update(df1['Image URL List'])

    df__.reset_index(inplace=True)
    df1.reset_index(inplace=True)
```

```python
def replace_empty_list_with_none(lst):
    if isinstance(lst, list) and len(lst) == 0:
        return None
    return lst

# Apply the custom function to the 'Image URL List' column
df_['Image URL List'] = df_['Image URL List'].apply(replace_empty_list_with_none)

df.set_index('MAL_ID', inplace=True)
df_.set_index('MAL_ID', inplace=True)

df.update(df_['Image URL List'])

df.reset_index(inplace=True)
df_.reset_index(inplace=True)

df.to_csv('step_3_ImageURL.csv', index=False)

df.isnull().sum()
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_image_urls(search_query):
    url = f"https://www.google.com/search?q={search_query}&tbm=isch"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
    
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.content, 'html.parser')
    image_tags = soup.find_all('img', class_='t0fcAb', limit=5)
    
    image_urls = []
    for image_tag in image_tags:
        image_url = image_tag.get('src')
        image_urls.append(image_url)
    
    return image_urls

# Example usage:
search_query = "cats"
image_urls = fetch_image_urls(search_query)
print(image_urls)
```
